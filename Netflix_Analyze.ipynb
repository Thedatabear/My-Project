```python
#create empty data frame
directors_list = pd.DataFrame()
print(directors_list)
```

```python
#split 1 comma text into 2 texts
directors_list = df['director'].str.split(',', expand=True).stack()
print(directors_list)
```

```python
#change director data to a list of director names one per row.
directors_list = directors_list.to_frame()
print(directors_list)
```

```python
#create columns for 'director list'
directors_list.columns = ['Director']
print(directors_list)
```

```python
# create a list of directors and their movies, counts how many movies each director has made
directors = directors_list.groupby(['Director']).size().reset_index(name='counts')
print(directors)
```

```python
# cleaning the data by removing a placeholder value ('Director not specified')
directors = directors[directors.Director != 'Director not specified']
print(directors)
```

```python
#rearranges the DataFrame directors to show the directors ranked by their movie counts
directors = directors.sort_values(by=['counts'], ascending=False)
print(directors)
```

```python
from textblob import TextBlob

df3 = df[['release_year', 'description']].copy()

df3.rename(columns={'release_year': 'Release Year', 'description': 'Description'}, inplace=True)


for index, row in df3.iterrows():
    d = row['Description']
    testimonial = TextBlob(str(d))
    p = testimonial.sentiment.polarity

    if p == 0:
        sent = 'Neutral'
    elif p > 0:
        sent = 'Positive'
    else:
        sent = 'Negative'

    df3.loc[index, 'Sentiment'] = sent

# Calculate the amount by year and sentiment
sentiment_summary = df3.groupby(['Release Year', 'Sentiment']).size().reset_index(name='Total Count')

# Filter> 2005
filtered_df = sentiment_summary[sentiment_summary['Release Year'] > 2005]

# Create a bar graph
import plotly.express as px
barGraph = px.bar(filtered_df,
                  x="Release Year",
                  y="Total Count",
                  color="Sentiment",
                  title="Sentiment Analysis of Netflix Content")
barGraph.show()
```
